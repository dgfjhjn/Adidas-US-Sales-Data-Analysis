```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('/content/Adidas US Sales Datasets.xlsx', skiprows=4)
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])
df['Year'] = df['Invoice Date'].dt.year
df['Month_Name'] = df['Invoice Date'].dt.month_name()
df['Day_Name'] = df['Invoice Date'].dt.day_name()

```

#من هم وكلاء البيع  الأفضل أداءً؟

```python
retailer_totals = df.groupby('Retailer')[['Total Sales', 'Units Sold', 'Operating Profit']].sum().reset_index().sort_values(by='Total Sales', ascending=False)

retailer_totals
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=retailer_totals, x='Retailer', y='Total Sales', palette='Blues_r')
plt.title('Total Sales by Retailer', fontsize=14, weight='bold')
plt.xlabel('Retailer', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

# التوزيع الجغرافي للمبيعات

```python
geo_totals = df.groupby(['Region', 'State'])[['Total Sales', 'Operating Profit']].sum().reset_index().sort_values(by='Total Sales', ascending=False)
geo_totals
```

```python
top_states = geo_totals.groupby('State')['Total Sales'].sum().reset_index().sort_values('Total Sales', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.hlines(y=top_states['State'], xmin=0, xmax=top_states['Total Sales'], color='skyblue', linewidth=3)
plt.plot(top_states['Total Sales'], top_states['State'], "o", markersize=10, color='darkblue')
plt.title('Top 10 States by Total Sales ', fontsize=14, weight='bold')
plt.xlabel('Total Sales melion ($)', fontsize=12)
plt.ylabel('State', fontsize=12)
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.show()
```

```python
heatmap_data = geo_totals.set_index('State')[['Total Sales', 'Operating Profit']].head(15)
plt.figure(figsize=(8, 8))
sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='YlGnBu', linewidths=0.5)
plt.title('Sales & Profit Concentration - Top 15 States\n', fontsize=14, weight='bold')
plt.ylabel('State', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

#هل الاعتماد على المتاجر اكثر  ولا الاونلاين

```python
method_totals = df.groupby('Sales Method')[['Total Sales', 'Operating Profit']].sum().reset_index().sort_values(by='Total Sales', ascending=False)
method_totals
```

```python
sns.set_theme(style="white")
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    data=method_totals,
    x='Sales Method',
    y='Total Sales',
    palette='mako',
    hue='Sales Method',
    legend=False
)

for container in ax.containers:
    ax.bar_label(container, fmt='${:,.0f}', padding=3, fontweight='bold')

sns.despine(left=True)
plt.yticks([])
plt.title('Total Sales by Method', fontsize=14, pad=20, fontweight='bold')
plt.xlabel('Sales Method', fontsize=12)
plt.show()
```

```python
custom_colors = ['#443963', '#4b7a95', '#5eb2a0']

fig = px.pie(method_totals,
             values='Total Sales',
             names='Sales Method',
             title='Modern Sales Distribution',
             color_discrete_sequence=custom_colors)
fig.update_traces(textposition='inside', textinfo='percent+label')

fig.show()
```

#المنتجات التي تحقق أعلى  ربح  بالنسبة لحجم مبيعاتها

```python
product_margin = df.groupby('Product')[['Total Sales', 'Operating Profit']].sum()
product_margin['Margin %'] = (product_margin['Operating Profit'] / product_margin['Total Sales']) * 100
product_margin = product_margin.sort_values('Margin %', ascending=False).reset_index()

product_margin
```

```python
fig1 = px.bar(product_margin, x='Margin %', y='Product', orientation='h',
             title='Profit Margin Efficiency by Product',
             color_discrete_sequence=['#4b7a95'])
fig1.show()
```

#قناة البيع الأكثر نجاحاً في كل منطقة جغرافية

```python
region_channel = df.groupby(['Region', 'Sales Method'])['Total Sales'].sum().reset_index()
region_channel.sort_values(by='Total Sales', ascending=False)
```

```python
fig3 = px.sunburst(region_channel, path=['Region', 'Sales Method'], values='Total Sales',
                  title='Regional Channel Synergy',
                  color_discrete_sequence=['#443963', '#4b7a95', '#5eb2a0'])
fig3.show()
```

```python
fig = px.bar(region_channel,
             x='Region',
             y='Total Sales',
             color='Sales Method',
             barmode='group',
             title='Sales Method Comparison by Region',
             color_discrete_sequence=['#443963', '#4b7a95', '#5eb2a0'], #
             labels={'Total Sales': 'Total Sales ($)', 'Region': 'Region'})
fig.update_layout(
    legend_title_text='Sales Method',
    hovermode='x unified'
)

fig.show()
```
